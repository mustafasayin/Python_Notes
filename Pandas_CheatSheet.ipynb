### loc & iloc
Gözlem ve değişken seçimi işlemleri: loc&iloc

```python
m = np.random.randint(1,30, size = (10,3))
df = pd.DataFrame(m,columns = ["var1", "var2", "var3"])
df
```

***loc: tanımlandığı şekil ile seçim yapmak için kullanılır.***
istenen indexten istenen indexe kadar seçim yapar. yani 0'dan 3'e kadar getirir(ikisi de dahil)

```python
df.loc[0:3]
```

***iloc: alışık olunan indeksleme ile seçim yapmak için kullanılır.***
index işlemlerine sadık bir şekilde getirir.

```python
df.iloc[0:3]
```

```python
df.iloc[:3,:2]
```

```python
df.loc[0:3, "var1"]
```

```python
df.iloc[0:4]["var1"]
```

#### İleri Gruplama (aggregate, filter, transform, apply)
**aggregate** fonksiyonu, bir DataFrame veya Series üzerinde birden fazla toplulaştırma işlemi uygulamak için kullanılır. Bu işlemler, veri sütunlarına uygulanacak farklı fonksiyonlar olabilir. Örneğin, mean, sum, min, max gibi fonksiyonlar.

```python
df.groupby("gruplar").mean()
```

```python
df.groupby("gruplar").aggregate(["min", np.median, max])
```

```python
df.groupby("gruplar").aggregate(["min", "max", "mean","sum","var"])
```

```python
df
```

```python
result = df.aggregate({'veri': 'sum', 'veri2': 'mean'})
print(result)
```

```python
result2 = df.aggregate({'veri': lambda x: x.max() - x.min(), 'veri2': 'std'})
print(result2)
```

```python
df = pd.DataFrame({"gruplar": ["A","B","C","A","B","C"],
                   "degisken1": [10,20,30,40,50,60],
                   "degisken2": [11,22,33,44,55,66]},
                  columns =["gruplar","degisken1","degisken2"])
df
```

```python
df.groupby("gruplar").aggregate(["min","max","sum","mean"])
```

```python
df.groupby("gruplar").aggregate({"degisken1": "min", "degisken2":"sum"})
```

FILTER

```python
def filter_func(x):
  return x["degisken1"].std()>9
```

```python
df.groupby("gruplar").filter(filter_func)
```

TRANSFORM

```python
df2 = df.iloc[:,1:3]
df2
```

```python
df.iloc[:,1:3].transform(lambda x: x-x.mean())
```

```python
df2.transform(lambda x: (x-x.mean())/x.std())
```

APPLY

```python
df.groupby("gruplar").apply(np.sum)
```

### Pivot Table
groupby'ın çok boyutlu hali gibi düşünülebilir.

```python
titanic = sns.load_dataset("titanic")
titanic
```

```python
titanic.groupby("embark_town")[["survived"]].mean()
```

```python
titanic.groupby(["sex","class"])[["survived"]].aggregate("mean").unstack()
```

```python
titanic.pivot_table("survived", index = "sex", columns="class" )
```

```python
age = pd.cut(titanic["age"], [0,18,60])
age
```

```python
titanic.pivot_table("survived", ["sex", age], "class")
```
